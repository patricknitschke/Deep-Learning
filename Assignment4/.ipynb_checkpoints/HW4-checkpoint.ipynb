## [TO COMPLETE] Exercise 4.1: Simple RNN

Let's start by importing Tensorflow, Keras and Numpy

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt

np.random.seed(42)
```

### Load dataset:
In this HW, we use the same datset used in the HW2, the IMDB dataset. The dataset contains 50,000 movie reviews from IMDB, labeled by sentiment (positive/negative). Reviews have been preprocessed, and each review is encoded as a sequence of word indexes. For convenience, the words are indexed by the overall frequency in the dataset, so that for example the integer "3" encodes the 3rd most frequent word in the data. For testing purposes, we will only consider the first 10,000  most common words.
By default, the load_data method returns a breakdown of the dataset into training and test sets. Both these sets contain 25,000 samples. To also have a validation set, we split the test set in half.

```python

num_words = 10000
(X_train, y_train), (X_test, y_test) = keras.datasets.imdb.load_data(num_words=num_words)
(X_valid, X_test) = X_test[:12500], X_test[12500:]
(y_valid, y_test) = y_test[:12500], y_test[12500:]
```

Let's also get the word indexs (word to word-id)

```python
word_index = keras.datasets.imdb.get_word_index()
```

Now we create a reverse index (word-id to word) method. Moreover, we add three special word-ids to encode:
- the padding;
- the start of a sequence;
- a word that is not in the vocabulary of the first 10,000 most common words.

Moreover, we also add an "unknown" placeholder for all the other symbols (not words) that may occur.
*Please, notice that Keras does not use index 0, so we can shift the indices only by 3 positions.*

```python
reverse_index = {word_id + 3: word for word, word_id in word_index.items()}
reverse_index[0] = "<pad>" # padding
reverse_index[1] = "<sos>" # start of sequence
reverse_index[2] = "<oov>" # out-of-vocabulary
reverse_index[3] = "<unk>" # unknown

def decode_review(word_ids):
    return " ".join([reverse_index.get(word_id, "<err>") for word_id in word_ids])

word_index = {word: word_id for word_id, word in reverse_index.items()}

rev = '''hello world and friends wow who would have thought that this thing actually works
was a very good movie and i like it a lot my favorite part is also good
and i would recommend this to all and watch it again everyday i am also very pleased
and now i will celebrate watch this amazing movie in 3d'''.split()
X_test[0] = np.array([word_index[word] if word_index[word] < num_words else 2 for word in rev])
y_test[0] = 1
```

Let's print a training sample and its target value

```python
decode_review(X_test[0])
```

```python
y_test[0]
```

Because of a limit of Keras, to feed the input data into an RNN model we have to create sequences that have the same length. We use keras.preprocessing.sequence.pad_sequences() to preprocess X_train: this will create a 2D array of 25,000 rows (one per review) and maxlen=500 columns. Because of that, reviews longer than 500 words will be cut, while reviews shorter than 500 words will be padded with zeros.

```python
maxlen = 500
X_train_trim = keras.preprocessing.sequence.pad_sequences(X_train, maxlen=maxlen)
X_test_trim = keras.preprocessing.sequence.pad_sequences(X_test, maxlen=maxlen)
X_valid_trim = keras.preprocessing.sequence.pad_sequences(X_valid, maxlen=maxlen)
```

### [TO COMPLETE] Define the model:
Let's define the model: 
- The first layer is an Embedding layer, with input_dim=num_words and output_dim=10. The model will gradually learn to represent each of the 10,000 words as a 10-dimensional vector. So the next layer will receive 3D batches of shape (batch size, 500, 10)
- The second layer is the recurrent one. In particular, in this case, we use a [SimpleRNN](https://www.tensorflow.org/api_docs/python/tf/keras/layers/SimpleRNN) 
- The output layer is a Dense layer with a sigmoid activation function since this is a binary classification problem

[TO COMPLETE] In the following cell, we already inserted in the model the first Embedding layer. Add the recurrent layer (using 32 units), and the output layer. Select the right activation function for the output layer and motivate your choice. Finally, select the right loss function inserting the right value for the "loss" parameter in model.compile() and motivate your choice.
Please, insert your answers in a new text cell below this one, immediately before the code.

#### Answer
The activation in the output layer should have an ouput from 0 to 1, since our we have a binary classification problem with targets 0 and 1. Thus, we can use the sigmoid activation function. For the loss function, we need one that calculates the loss for only two targets, and so we choose the BinaryCrossEntropy loss function.

```python
model = keras.models.Sequential()
model.add(keras.layers.Embedding(input_dim=num_words, output_dim=10))
model.add(keras.layers.SimpleRNN(units=32))
model.add(keras.layers.Dense(units=1, activation="sigmoid"))

model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

Let's print a summary of the model. Specifically, note the number of parameters of the RNN layer.

```python
model.summary()
```

### Train the model:
Now we have to train the model

```python
history = model.fit(X_train_trim, y_train,
                    epochs=10, batch_size=128, validation_data=(X_valid_trim, y_valid))
```

Print the values of accuracy and the loss , and evaluate the model on the test set

```python
def plot_loss(history):
    plt.figure(figsize=(10,6))
    plt.plot(history.epoch,history.history['loss'])
    plt.plot(history.epoch,history.history['val_loss'])
    plt.title('loss')

def plot_accuracy(history):
    plt.figure(figsize=(10,6))
    plt.plot(history.epoch,history.history['accuracy'])
    plt.plot(history.epoch,history.history['val_accuracy'])
    plt.title('accuracy')

plot_loss(history)
plot_accuracy(history)

scores = model.evaluate(X_test_trim, y_test, verbose=2)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
y = model.predict(X_test_trim)
y = y[0]
input_text= X_test_trim[0]

print(decode_review(input_text))
print("Ouput:", y)
```

### LSTM Model

```python
model_lstm = keras.models.Sequential()
model_lstm.add(keras.layers.Embedding(input_dim=num_words, output_dim=10))
model_lstm.add(keras.layers.LSTM(units=14))
model_lstm.add(keras.layers.Dense(units=1, activation="sigmoid"))

model_lstm.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
model_lstm.summary()

history_lstm = model_lstm.fit(X_train_trim, y_train,
                    epochs=10, batch_size=128, validation_data=(X_valid_trim, y_valid))
```

```python
plot_loss(history_lstm)
plot_accuracy(history_lstm)

scores_lstm = model_lstm.evaluate(X_test_trim, y_test, verbose=2)
print("%s: %.2f%%" % (model_lstm.metrics_names[1], scores_lstm[1]*100))
```

### GRU Model

```python
model_gru = keras.models.Sequential()
model_gru.add(keras.layers.Embedding(input_dim=num_words, output_dim=10))
model_gru.add(keras.layers.GRU(units=16))
model_gru.add(keras.layers.Dense(units=1, activation="sigmoid"))

model_gru.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
model_gru.summary()

history_gru = model_gru.fit(X_train_trim, y_train,
                    epochs=10, batch_size=128, validation_data=(X_valid_trim, y_valid))
```

```python
plot_loss(history_gru)
plot_accuracy(history_gru)

scores_gru = model_gru.evaluate(X_test_trim, y_test, verbose=2)
print("%s: %.2f%%" % (model_gru.metrics_names[1], scores_gru[1]*100))
```

### Discussion

First, we notice that for both the LTSM and GRU models, the number of units have to be decreased significantly in order to match the number of parameters in the standard RNN model. This is most likely due to gates in the LTSM model and the attention in GRU model, which increases the number of weights to be learnt.

In regards to performance and training time, we observe that the performance increases for both the LTSM and GSU models, albeit overfitting at the end of the training period as well. Yet, for this slight performance increase, there is also an increase in training times for both models, where training times almost double. Also, we note that the LTSM, with roughly the same training times as the GRU model, performs the best. It is difficult to say exactly what the exact reason for the performance increase is, but we can argue that the training time difference could be explained by the number of operations and extra activations/gradients in the gates that have to be computed in each step.

## Bidirectional LSTM
Let's modify the previous code by using a bidirectional LSTM instead of a simple LSTM. In Keras, it is possible to define a bidirectional layer by using [tf.keras.layers.Bidirectional](https://www.tensorflow.org/api_docs/python/tf/keras/layers/Bidirectional). Note that this wrapper requires as argument a layer, in our case we use [tf.keras.layers.LSTM](https://www.tensorflow.org/api_docs/python/tf/keras/layers/LSTM)


```python
model_bidirectional = keras.models.Sequential()
model_bidirectional.add(keras.layers.Embedding(input_dim=num_words, output_dim=10))
model_bidirectional.add(keras.layers.Bidirectional(keras.layers.LSTM(32)))
model_bidirectional.add(keras.layers.Dense(1, activation="sigmoid"))

model_bidirectional.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

model_bidirectional.summary()

history = model_bidirectional.fit(X_train_trim, y_train,
                    epochs=5, batch_size=128, validation_data=(X_valid_trim, y_valid))
```

```python
plot_loss(history)

plot_accuracy(history)

scores = model_bidirectional.evaluate(X_test_trim, y_test, verbose=2)
print("%s: %.2f%%" % (model_bidirectional.metrics_names[1], scores[1]*100))
```

### Dropout
A droupout rate that reduces the overfitting issue quite well is *rate = 0.8*, when applied to both the bidirectional and embedded layers.

```python
drop_models = {}
drop_histories = {}
```

```python
rates = [0.8]
for rate in rates:
    model_bidirectional = keras.models.Sequential()
    model_bidirectional.add(keras.layers.Embedding(input_dim=num_words, output_dim=10))
    model_bidirectional.add(keras.layers.Dropout(rate=rate))
    model_bidirectional.add(keras.layers.Bidirectional(keras.layers.LSTM(32)))
    model_bidirectional.add(keras.layers.Dropout(rate=rate))
    model_bidirectional.add(keras.layers.Dense(1, activation="sigmoid"))

    model_bidirectional.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
    model_bidirectional.summary()

    history = model_bidirectional.fit(X_train_trim, y_train,
                        epochs=5, batch_size=128, validation_data=(X_valid_trim, y_valid))
    
    drop_histories[rate] = history
    drop_models[rate] = model_bidirectional
```

```python
for rate, history in drop_histories.items():
    plot_loss(history)
    plot_accuracy(history)

    scores = model_bidirectional.evaluate(X_test_trim, y_test, verbose=2)
    print(f"Dropout rate: {rate},","%s: %.2f%%" % (model_bidirectional.metrics_names[1], scores[1]*100))
```

### Experimental study
It is expected that the hyperparameters such as the:
1. number of units in a layer
2. number of epochs
3. batch size
4. learning rate

can have an effect on the degree of overfitting in the model. Of this, we know that many units lead to a higher complexity and ability to overfit the data. Similarly, a large number of epochs allow the model to fit the data a lot more than necessary, which can also lead to overfitting. Next, we also expect that a smaller batch size will introduce a sort of regularisation due to noise. Lastly, to reduce overfitting we can try to reduce the learning rate in attempt to not "too quickly" converge to a suboptimal solution, yet this we be left untested.

To test these assumptions, we can attempt to run some experiments where we *lower* the number of units and *decrease* the batch size. With only 5 epochs already, reducing this will probably not allow the model to train adequately, so we will leave this as it is.

```python
models = {}
model_histories = {}
```

```python
batch_sizes = [128, 32]
units = [32, 8]

for n_units in units:
    for batch in batch_sizes:
        
        model_bidirectional = keras.models.Sequential()
        model_bidirectional.add(keras.layers.Embedding(input_dim=num_words, output_dim=10))
        model_bidirectional.add(keras.layers.Bidirectional(keras.layers.LSTM(n_units)))
        model_bidirectional.add(keras.layers.Dense(1, activation="sigmoid"))

        model_bidirectional.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
        model_bidirectional.summary()

        history = model_bidirectional.fit(X_train_trim, y_train,
                            epochs=5, batch_size=batch, validation_data=(X_valid_trim, y_valid))
        key = f"n_units : {n_units}, batch_size: {batch}"
        models[key] = model_bidirectional
        model_histories[key] = history
```

```python
for key, history in model_histories.items():
    plot_loss(history)
    plot_accuracy(history)

    scores = models[key].evaluate(X_test_trim, y_test, verbose=2)
    print(f"Model params: {key},","%s: %.2f%%" % (model_bidirectional.metrics_names[1], scores[1]*100))
```

From this quick experiment, we see that the batch size (Graph 2 and 4) does not contribute significantly to a reduce in overfitting. We do however, observe that by reducing the number of inputs in the bidirectional LTSM layer from 32 (Graphs 1 and 2) units to 8 (Graphs 3 and 4), we do get a significant decrease of overfitting.

Of course, we may attempt other regularisation techniques or a combination of these to mitigate overfitting. But in regards to the question posed, *if it is possible to mitigate overfitting by means other than dropout*, then the answer is yes. In this case, we see that by simply reducing the complexity of the model through a reduced number of units - we obtain less overfitting, though not necessarily better than adding dropout.

```python
# Test for fun
model = models[f"n_units : {8}, batch_size: {128}"]
y = model.predict(X_test_trim)
y = y[0]
input_text= X_test_trim[0]

print(decode_review(input_text))
print("Ouput:", y)
```
